```python
def limpeza1(linha):
    resultado = str(linha["nome_uf"]).strip()
    if resultado == "Minas Gerais":
        retorna = "MG"
    elif resultado == "São Paulo":
        retorna = "SP"
    elif resultado == "Rio Grande do Sul":
        retorna = "RS"
    elif resultado == "Bahia":
        retorna = "BA"
    elif resultado == "Paraná":
        retorna = "PR"
    elif resultado == "Santa Catarina":
        retorna = "SC"
    elif resultado == "Goiás":
        retorna = "GO"
    elif resultado == "Piauí":
        retorna = "PI"
    elif resultado == "Paraíba":
        retorna = "PB"
    elif resultado == "Maranhão":
        retorna = "MA"
    elif resultado == "Pernambuco":
        retorna = "PE"
    elif resultado == "Ceará":
        retorna = "Ceará"
    elif resultado == "Rio Grande do Norte":
        retorna = "RN"
    elif resultado == "Pará":
        retorna = "PA"
    elif resultado == "Mato Grosso":
        retorna = "MT"
    elif resultado == "Tocantins":
        retorna = "TO"
    elif resultado == "Alagoas":
        retorna = "AL"
    elif resultado == "Rio de Janeiro":
        retorna = "RJ"
    elif resultado == "Mato Grosso do Sul":
        retorna = "MS"
    elif resultado == "Espírito Santo":
        retorna = "ES"
    elif resultado == "Sergipe":
        retorna = "SE"
    elif resultado == "Amazonas":
        retorna = "AM"
    elif resultado == "Rondônia":
        retorna = "RO"
    elif resultado == "Acre":
        retorna = "AC"
    elif resultado == "Amapá":
        retorna = "AP"
    elif resultado == "Roraima":
        retorna = "RR"
    elif resultado == "Distrito Federal":
        retorna = "DF"
    
    return retorna

conta_geral_municipios["sigla_uf"] = conta_geral_municipios.apply(limpeza1, axis=1)
```

```python
def encontra(a, b):
  return not set(a).isdisjoint(b)
```

```python
for num, row in todos_anos.iterrows():
    consistencia = row['consistencia']
    pt_monitoramento = row['ponto_de_monitoramento']
    atendimento_ao_padrao = row['atendimento_ao_padrao']
    ano = row['ano_de_referência']
    semestre = row['semestre_de_referência']

    grupo = row['grupo_de_parâmetros']
    
    substancia = row['parâmetro']
    if substancia in dict_subst:
       substancia_inteira = dict_subst[substancia]

    substancia_separada = substancia_inteira.split("-")[0].strip().upper()

    if grupo == 'Parâmetros Organolépticos':
      dicionario = {"cod_ibge": str(row['código_ibge']).strip(), 
                    "uf": str(row['uf']).strip(), 
                    "ano": ano, 
                    "semestre": semestre, 
                          "substancia": substancia_inteira.strip(), 
                          "grupo": grupo.strip(), 
                    "consistencia": consistencia,
                          "pt_monitoramento": pt_monitoramento,
                          "atendimento_ao_padrao": atendimento_ao_padrao,
                          "tipo_cor": "8"
                          }
    elif atendimento_ao_padrao == 'Inconclusivo' or atendimento_ao_padrao == 'N/A' or atendimento_ao_padrao == 'not applicable' or atendimento_ao_padrao == '' or pd.isna(atendimento_ao_padrao) :
      dicionario = {"cod_ibge": str(row['código_ibge']).strip(), 
                    "uf": str(row['uf']).strip(), 
                    "ano": ano, 
                    "semestre": semestre, 
                          "substancia": substancia_inteira.strip(), 
                          "grupo": grupo.strip(), 
                          "consistencia": consistencia,
                          "pt_monitoramento": pt_monitoramento,
                          "atendimento_ao_padrao": atendimento_ao_padrao,
                          "tipo_cor": '6'
                          }
    elif (substancia_separada in subst) and (consistencia == 'Consistente' and pt_monitoramento != 'PONTO DE CAPTAÇÃO') and atendimento_ao_padrao == 'Acima do VMP': 
      dicionario = {"cod_ibge": str(row['código_ibge']).strip(), 
                    "uf": str(row['uf']).strip(), 
                    "ano": ano, 
                    "semestre": semestre, 
                          "substancia": substancia_inteira.strip(), 
                          "grupo": grupo.strip(), 
                    "consistencia": consistencia,
                          "pt_monitoramento": pt_monitoramento,
                          "atendimento_ao_padrao": atendimento_ao_padrao,
                          "tipo_cor": '1'
                          }
    elif (substancia_separada not in subst) and (consistencia == 'Consistente' and pt_monitoramento != 'PONTO DE CAPTAÇÃO') and atendimento_ao_padrao == 'Acima do VMP': 
     dicionario = {"cod_ibge": str(row['código_ibge']).strip(), 
                   "uf": str(row['uf']).strip(), 
                   "ano": ano, 
                    "semestre": semestre, 
                          "substancia": substancia_inteira.strip(), 
                          "grupo": grupo.strip(), 
                   "consistencia": consistencia,
                          "pt_monitoramento": pt_monitoramento,
                          "atendimento_ao_padrao": atendimento_ao_padrao,
                          "tipo_cor": '2'
                          }
    elif (substancia_separada in subst) and (consistencia == 'Consistente' and pt_monitoramento != 'PONTO DE CAPTAÇÃO') and atendimento_ao_padrao == 'Abaixo do VMP': 
     dicionario = {"cod_ibge": str(row['código_ibge']).strip(), 
                   "uf": str(row['uf']).strip(), 
                   "ano": ano, 
                    "semestre": semestre, 
                          "substancia": substancia_inteira.strip(), 
                          "grupo": grupo.strip(), 
                   "consistencia": consistencia,
                          "pt_monitoramento": pt_monitoramento,
                          "atendimento_ao_padrao": atendimento_ao_padrao,
                          "tipo_cor": '3'
                          }
    elif (substancia_separada not in subst) and (consistencia == 'Consistente' and pt_monitoramento != 'PONTO DE CAPTAÇÃO') and atendimento_ao_padrao == 'Abaixo do VMP': 
     dicionario = {"cod_ibge": str(row['código_ibge']).strip(), 
                   "uf": str(row['uf']).strip(), 
                   "ano": ano, 
                    "semestre": semestre, 
                          "substancia": substancia_inteira.strip(), 
                          "grupo": grupo.strip(), 
                   "consistencia": consistencia,
                          "pt_monitoramento": pt_monitoramento,
                          "atendimento_ao_padrao": atendimento_ao_padrao,
                          "tipo_cor": '4'
                          }
  
    else:
      dicionario = {"cod_ibge": str(row['código_ibge']).strip(), 
                    "uf": str(row['uf']).strip(), 
                    "ano": ano, 
                    "semestre": semestre, 
                          "substancia": substancia_inteira.strip(), 
                          "grupo": grupo.strip(), 
                    "consistencia": consistencia,
                          "pt_monitoramento": pt_monitoramento,
                          "atendimento_ao_padrao": atendimento_ao_padrao,
                          "tipo_cor": "7"
                          }
  
    cidades.append(dicionario)
```

```python
alerta_max = df_cidades_novo[(df_cidades_novo['tipo_cor'] == '1') |
           (df_cidades_novo['tipo_cor'] == '2')].groupby(['cod_ibge', 'substancia', 'tipo_cor', 'ano']).size().sort_values(ascending=False).reset_index()
alerta_max.columns = ['cod_ibge', 'substancia', 'tipo_cor', 'ano', 'contagem']  
```

```python
# Cria dataframe com cod_ibge de tipos 1 e 2 
aparicoes_max = alerta_max.drop_duplicates(['cod_ibge'], keep='first')
aparicoes_max = aparicoes_max[['cod_ibge']]
```

```python
# Itera em cada municipio
dados_encontrados = []
for num, row in aparicoes_max.iterrows():
    codigo = row['cod_ibge']

    # Filtra os dados 
    filtra = alerta_max[(alerta_max['cod_ibge'] == codigo)]

    # Cria uma lista com as substancias
    arr = filtra["substancia"].unique()
    lista = arr.tolist()

    # Verifica se cada substancia apareceu tres anos - itera nas substancias
    for sub in lista:
      # Filta conjunto da substancia
      procura_ano = filtra[(filtra['substancia'] == sub)]
      
      # Separa que tipo_cor são
      tipo = procura_ano["tipo_cor"].unique()
      tipo = tipo.tolist()
      tipo = ''.join(tipo)

      # Cria lista com os anos
      arr2 = procura_ano["ano"].unique()
      anos_encontrados = arr2.tolist()

      # Verifica se os três anos aparecem na mesma substancia
      if '2020' in anos_encontrados and '2018' in anos_encontrados and '2019' in anos_encontrados:
        dicionario = {"cod_ibge": codigo, 
                  "substancia": sub,
                  "tipo_cor": tipo}
        dados_encontrados.append(dicionario) 
```

```python
df_alerta_max = pd.DataFrame(dados_encontrados)
df_alerta_max
```

```python
df_alerta_max = pd.merge(df_alerta_max, nomes_cidades, 
                                    how = 'left',
                       left_on = 'cod_ibge', 
                       right_on = 'id_municipio_6')

df_alerta_max = df_alerta_max[['cod_ibge', 'nome_municipio_oficial','substancia', 'tipo_cor']]
```
